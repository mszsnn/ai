```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 读取 CSV
data = pd.read_csv('people-example.csv')

# 2. 快速探索
print(data.head())
print(data.describe())

# 3. 可视化
sns.histplot(data['age'])
plt.show()

sns.scatterplot(x='age', y='salary', data=data)
plt.show()
```
